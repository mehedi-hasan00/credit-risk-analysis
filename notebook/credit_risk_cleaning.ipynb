```python
df['person_age'].value_counts()
```

```python
# we have to drop ages that are greater than 100 years
df[df['person_age'] > 100]
```

```python
df = df[df['person_age'] <= 100]
```

```python
df['person_emp_length'].value_counts()
```

```python
df = df[(df['person_emp_length'].isnull()) | (df['person_emp_length'] < 60)]
```

```python
df['person_emp_length'] = df['person_emp_length'].fillna(df['person_emp_length'].median())
```

```python
df['person_emp_length'].isnull().sum()
```

```python
df['loan_int_rate'].describe()
```

```python
df['loan_int_rate'] = df.groupby('loan_grade')['loan_int_rate'].transform(
    lambda x: x.fillna(x.median())
)
```

```python
df.isnull().sum()
```

```python
df['person_income'].quantile(0.99)
```

```python
# fresher and parttime job holders imcomes are less than other so i use only upper bounds to handle outliers
upper = df['person_income'].quantile(.99)
df['person_income'] = df['person_income'].clip(upper=upper)
```

```python
df.describe()
```

```python
df.dtypes
```

```python
df.head()
```

```python
df['loan_status_label'] = df['loan_status'].replace({0: "Non-Default",
                                                1: "Default"})
```

```python
df.tail()
```

```python
labels = ['20-25','26-30','31-40','41-50','50+']
df['age_group'] = pd.cut(df['person_age'], bins= [20,25,30,40,50,100], labels=labels)
```

```python
df['person_income'].value_counts(bins=5)
```

```python
bins=[0,48000,92000,136000,225000]
labels = ['Low','Medium','High','Very High']
df['income_group'] = pd.cut(df['person_income'], bins = bins, labels= labels)
```

```python
df['loan_amnt'].value_counts(bins = 5)
```

```python
bins = [0,7400, 14300, 21200, 35000]
labels = ['Small','Medium','Large','Very Large']
df['loan_group'] = pd.cut(df['loan_amnt'], bins=bins, labels= labels)
```

```python
df.head()
```

```python
df.rename(columns={
    'person_age':'age',
    'person_income':'income',
    'person_home_ownership':'home_ownership',
    'person_emp_length':'emp_length',
    'loan_amnt' : 'loan_amount',
    'loan_int_rate':'interest_rate',
    'loan_percent_income':'loan_to_income',
    'cb_person_default_on_file':'prev_default',
    'cb_person_cred_hist_length':'credit_hist_length'
}, inplace=True)
```

```python
df.head()
```

```python
df['prev_default'] = df['prev_default'].map({'Y':'Yes',
                                            'N':'No'})
```

```python
df['loan_risk'] = df['loan_grade'].map({
    'A': 'Low Risk',
    'B': 'Low Risk', 
    'C': 'Medium Risk',
    'D': 'Medium Risk',
    'E': 'High Risk',
    'F': 'High Risk',
    'G': 'High Risk'
})
```

```python
df['interest_rate_group'] = pd.cut(df['interest_rate'],
    bins=[0, 8, 12, 16, 25],
    labels=['Low', 'Medium', 'High', 'Very High'])
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['age_group'])
```

```python
df.to_csv('credit_risk_cleaned.csv', index=False)
```
